### Column Printer Function

Define and call two functions to print all the rows a specified column. This is because imported data can be viewed without looking at Excel, and any changes to the written data in columns can be reviewed without needing to save a file.

The functions should accept 2 arguments:
* Worksheet where data is located
* Column letter of data to print

The first function, cp1:
* Print cell coordinate (e.g. A1 to A28).
* Print Order ID (column A in sheet, 'Orders_Info')

The second function, cp2:
* Print cell content
* Print subtotal (column D in sheet, 'Orders_Info')

```python
# Provided information:
# The first function, cp1
# Loop through specified sheet and column - through cells with range via the .max_row method
# Use f-strings to navigate to cells
def cp1(sheet, col):
    for i in range(1, sheet.max_row + 1):
        print(f'{col}{i}', sheet[f'{col}{i}'].value)

cp1(orders, 'A')
```

```python
# Provided information:
# The second function, cp2
# Loop through specified sheet and column - looping through column
# Use f-strings to navigate to cells via sheet['coordinate']
def cp2(sheet, col):
    for c in sheet[col]:
        print(f'{c.coordinate}', sheet[c.coordinate].value)
cp2(orders, 'D')
```

### Step 1: Define the function
Based on the modified dictionary, od (see immediate above), define a function to retrieve and sum any one of the three category values generated from function (taxcal) (i.e. sum of all pretaxes or sum of all tax or sum of all totals).

Note:
* Use index [2] to first access any of the three numbers. Then, combine with index [0], [1] or [2] for the respective category values
* The function should accept the two arguments (the index from above and the dictionary name, which is od in this case).
* Round the sum of values to 2 decimal places.

```python
# Step 1: Define the function as described. Not display ouput at this stage

def sum_category(index, od):

    total_sum = sum(item[2][index] for item in od.values())
    return round(total_sum, 2)
```

Call the function from step 1, apply the correct index to calculate the sum of:

```python
# Step 2:

pretax_sum = sum_category(0, od)

tax_sum = sum_category(1, od)

total_sum = sum_category(2, od)

print(f"Sum of all pretaxes: {pretax_sum:.2f}") # another method: .2f also rounds the value off to 2 dec 😸
print(f"Sum of all taxes: {tax_sum:.2f}")
print(f"Sum of all totals: {total_sum:.2f}")

```

### Step 6: Sales By Location
Calculate the sum of pretaxes by location.

Create a new empty dictionary to hold the information where location is the key, and revenue for that location is the value:
* Loop through dictionary values.
* Set a variable, named loc, to represent the location column.
* Create a new entry and set its value to zero if a location is not present in the new dictionary.
* Accumulative addition of the pretax value every time an iteration matches the location.

Example display output: 
{'Sun Valley': (sum of pretax for Sun Valley), 'Stowe': (sum of pretax for Stowe), 'Mammoth': (sum of pretax for Mammoth),}

```python

s = {}

for row in orders.iter_rows(min_row=2): #min_row argument is used to skip the header 😸
    loc = row[4].value
    pretax = row[3].value

    if loc not in s:
        s[loc] = 0

    s[loc] += pretax

pprint(s)
```

### Step 7: Accumulative Function
Define a function (name it as 'acc') to generalise calculating a sum of a column, grouped by catogorical values from another column (such as dates or ids).

The function should accept the following arguments:
* a: for passing in the index of the category column.
* b: for passing in the value to sum ([0] for pretax, [1] for tax & [2] for total)
* c: for passing in the name of the dictionary that holds the data (in this case, it will be 'od').

Note:
1. Create a local variable to hold a new, empty dictionary
2. Include the sums of each category into the empty dictionary
3. Return the local variable to output the dictionary when completed

```python


def acc(a, b, c):
    result = {}
    for key, value in c.items():
        category = value[a] 
        total_value = value[b][2] 

        if category not in result:
            result[category] = 0

        result[category] += total_value

    return result
```
